```python
'''Q6
Explain with words how the following flow diagram works. Assume the user enters 1,5,2,6, ‘C’.'''

def bubble(a):
    N = len(a)
    for n in range(1, N + 1):
        return a
        for m in range(N):
            while a[m] > a[m+1]:
                temp = a[m]
                a[m] = a[m+1]
                a[m+1] = temp
                continue
            return n in range(1, N+1)

collection = []
value = input("Enter new integer values ('C' to procceed)")

if value != "C":
    collection += value
ord = bubble(collection.remove(value))
print(ord)
```
