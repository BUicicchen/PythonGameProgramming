```python
'''This program randomly generates dice rolling operation.
It counts the occurrences of getting a 6 on a dice out of 1000 rolls, and calculate the probability of 6 appearing.
It also compares the expected probability with the probability calculated. '''

import random

count = 0
```

```python
for i in range(1000):
    number = random.randint(1,6)
    print(i + 1 , number)
    if number == 6:
        count = count + 1
```

```python
probability = (count / 1000) * 100
probability_expected = (1/6) * 100
```

```python
print("6 appears " + str(number) + " times")
print("The probably of 6s that this random dice rolling operator calculates is: " + str(probability) + "%")
print("I expect to get " + str(probability_expected) + "%")
```

```python
if probability_expected == probability:
    print("I expect this value.")
else:
    print("I didn't expect this value.")
```
