# My first scientific app in Python with Numpy

In this first example we will create a sine function and plot it. The equation is: $y = 3sin(2\pi x)$, where $x \in [-1,1]$

(1) Let's import the Numpy library

```python
import numpy as np

# -- Create the values for x. This is a vector (a.k.a list)
x = np.linspace(-1,1,1000)
y = 3 * np.sin(2*np.pi*x)
```

(2) Import the graphic library and show the function y

```python
import matplotlib.pyplot as plt
%matplotlib inline
# this is needed to show graphs inside the notebook
plt.plot(x,y)
```

```python
# (3) Calculate the differential of $y$ and show it.
dy_dx = np.diff(y)
plt.plot(x[:-1], dy_dx)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x$', fontsize = 14)
plt.title('$y=sin(2\pi x)$', fontsize = 14)
plt.savefig('save.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
plt.plot(t,s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.grid(True)
plt.savefig('test.png')
plt.show()
999
```

# Fun with statistics: app No. 2

Let's create a random sample of 1000 points from a Poisson distribution

```python
p = np.random.poisson(5, 1000)
plt.plot(p, color = 'r')
plt.grid()
plt.xticks(fontsize = 14) # numbers in the x axis
plt.yticks(fontsize = 14) # numbers in the y axis
```

```python
mu = np.mean(p, axis=0)
print('The expected value (mean) of the Poisson distribution is {}'. format(mu))
```

```python
b, c, c = plt.hist(p, 15, normed=True ) # 15 is the number of bars, also called bins
# b, c, and d are: the count, the buns, and the ignored values
```

```python
print('The total probabaility {} is'.format(sum(b)))
```

```python
cdf = np.cumsum(b)
plt.plot(cdf, color='black', linewidth = 3)
plt.xlim([0, 10])
plt.xlabel('Random number $x$', fontsize=14)
plt.ylabel('Commulative distribution', fontsize=14)
```

```python
## Graph for normal distribution ##

y = np.random.normal(loc = 86, scale=10, size=1000)
plt.plot(y, color='black', linewidth = 3)
plt.grid()
mu = np.mean(y, axis=0)

print('The expected value (mean) of the normal distribution is {}'. format(mu))
```
