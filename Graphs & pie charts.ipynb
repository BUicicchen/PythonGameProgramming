# Working with subplots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1,1000) # this creates an array of numbers from 0 to 1 with 1000 points
y = np.cos(x**2*np.pi*10)
```

```python
plt.plot(x,y) # traditional way of plotting using one window
```

```python
# now using subplots
f, axis = plt.subplots()
axis.plot(x,y,color='r')
axis.set_title('$y=sin(10/pi x^2)$')
```

```python
f, axis2 = plt.subplots(2) # define the container of the graphs
axis2[0].plot(x,y)
axis2[1].plot(x,[1/(i+0.001) for i in y], color='b')
```

```python
f, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x,y)
ax1.set_title('Sharing y-axis')
ax2.plot(x, [y[i] for i in range(len(y)-1, -1, -1)], color='g')
ax2.set_title('Reversed y')
f, (ax3, ax4) = plt.subplots(1,2) # define the container of the graphs
ax3.plot(x, [i**0.5 for i in y], color='y')
ax3.set_title('Square root of y')
ax4.plot(x, [abs(i) for i in y], color='r')
ax4.set_title('Absolute value of y')
```

# Population of students at PAS

```python
age = np.random.normal(14, 2.5, 500)
height = np.random.normal(150, 5, 500)
plt.scatter(age, height, color='r')

fig = plt.figure()
ax = fig.add_subplot(1, 2, 1)
ax.scatter(age, height, color='r')

ax = fig.add_subplot(1, 2, 2)
ax.hist(height, color='r', orientation='horizontal')

ax = fig.add_subplot(3,2,5)
ax.plot(np.sort(age),color='r')
```

# Pie chart in Python

```python
pets = ['dog', 'cats', 'horses', 'sea horses', 'unicorns']
freq = [10, 2, 20, 10, 55]

fig1, ax1 = plt.subplots() # container
ax1.pie([f/sum(freq) for f in freq], labels=pets, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') # equal aspect ratio ensures that pie is drawn
```
